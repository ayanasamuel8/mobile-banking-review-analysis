### load data

```python
import pandas as pd
df = pd.read_csv('../data/Cleaned/reviews_cleaned.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# sentiment distribution

```python
sns.countplot(data=df, x='sentiment_label')
plt.title("Overall Sentiment Distribution")
plt.show()
```

# rating distribution

```python
sns.histplot(df['rating'], bins=5, kde=True)
plt.title("Rating Distribution")
plt.show()
```

# word cloud for top positive and negative

```python
from wordcloud import WordCloud

positive_text = ' '.join(df[df['sentiment_label'] == 'positive']['review_text_clean'])
WordCloud(width=800, height=400).generate(positive_text).to_image()
```

```python
positive_text = ' '.join(df[df['sentiment_label'] == 'negative']['review_text_clean'])
WordCloud(width=800, height=400).generate(positive_text).to_image()
```

# top theme

```python
df['identified_theme'].value_counts().plot(kind='bar')
plt.title("Most Common Themes")
plt.show()
```
